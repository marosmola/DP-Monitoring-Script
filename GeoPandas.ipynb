```python
import json
from geopandas import GeoDataFrame
from geojson import FeatureCollection

with open('shapes/hlavna.geojson', 'r') as data:
    hlavna_data = json.loads(data.read())

with open('shapes/in.geojson', 'r') as data:
    in_data = json.loads(data.read())

with open('shapes/out.geojson', 'r') as data:
    out_data = json.loads(data.read())
    
    
feature1 = FeatureCollection(hlavna_data)
feature2 = FeatureCollection(in_data)
feature3 = FeatureCollection(out_data)

hlavna = GeoDataFrame.from_features(feature1)
point_in = GeoDataFrame.from_features(feature2)
point_out = GeoDataFrame.from_features(feature3)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

hlavna.plot(ax=ax, color='white', edgecolor='black')
point_in.plot(ax=ax, marker='o', color='red', markersize=5)
point_out.plot(ax=ax, marker='o', color='blue', markersize=5)

print(hlavna.intersects(point_in))
print(hlavna.intersects(point_out))

```
